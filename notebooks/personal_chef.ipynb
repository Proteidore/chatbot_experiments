```python
from dotenv import load_dotenv

load_dotenv()
```

```python
# agent parameters
model = 'gpt-5-nano'
system_prompt = "You are a smart cooker. Your task is to return a recipe when a list of ingredients is available. You can deduce also deduce a list of ingredients from a picture."
```

```python
# agent web search tool
from langchain.tools import tool
from typing import Dict, Any
from tavily import TavilyClient 

tavily_client = TavilyClient()

@tool
def web_search(query: str) -> Dict[str, Any]:

    """Search the web for information"""

    return tavily_client.search(query)
```

```python
# agent creation
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver  

agent = create_agent(
    model=model,
    system_prompt=system_prompt,
    tools = [web_search],
    checkpointer=InMemorySaver()
)
```

```python
# agent memory id
config = {"configurable": {"thread_id": "1"}}
```

```python
# Personal Chef chatbot
from ipywidgets import FileUpload
from IPython.display import display
import base64
from langchain.messages import HumanMessage

print("Here your personal chef ! Ask me a recipe from a photograph.")

while True:
    # load a picture if wanted
    uploader = FileUpload(accept='.png', multiple=False)
    display(uploader)

    # user request
    request = input()

    # word breaking the while-loop
    if request == "STOP" :
        break


    # creation of the content of the request sent to the model

    ## text
    content = [{"type": "text", "text": request}]

    ## picture
    if uploader.value:
        uploaded_file = uploader.value[0]
        content_mv = uploaded_file["content"]
        img_bytes = bytes(content_mv)
        img_b64 = base64.b64encode(img_bytes).decode("utf-8")

        content.append({
            "type": "image",
            "base64": img_b64,
            "mime_type": "image/*" #tout type d'images (jpg, png, ...)
        })

    multimodal_question = HumanMessage(content=content)

    # request execution
    response = agent.invoke(
    {"messages": [multimodal_question]},
    config)

    print(response['messages'][-1].content)
    print("")
```
